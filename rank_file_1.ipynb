```python
def classify(row,High_mean,medium_mean,low_cv,medium_cv):
    if row['Mean'] >= High_mean and row['CV'] <= low_cv:  # high & consistent
        return 'Fast'
    elif row['Mean'] >= medium_mean and row['CV'] <= medium_cv:
        return 'Normal'
    else:
        return 'Slow'
```

```python
df['Rank'] = df.apply(classify,axis=1,High_mean= High_mean,medium_mean=medium_mean,low_cv=low_cv,medium_cv=medium_cv)
```
